```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([500, 700, 900, 1100, 1300, 1500])
y = np.array([50, 65, 80, 95, 115, 130])
```

```python
print(f"X: {x}")
print(f"Y: {y}")
```

```python
plt.scatter(x, y)

plt.xlabel("House Size (sqft)")
plt.ylabel("Price (lakhs)")
plt.title("House Price vs Size")

plt.show()
```

```python
w = 0
b = 0

def predict(w, x, b):
    return w * x + b
```

```python
def compute_cost(w, x, b, y):

    y_pred = predict(w, x, b)

    m = len(x)

    cost = (1/(2 * m)) * np.sum((y_pred - y) ** 2)

    return cost
```

```python
cost = compute_cost(w, x, b, y)
print(cost)
```

```python
def compute_gradient(w, x, b, y):

    m = len(x)

    y_pred = predict(w, x, b)

    dj_dw = (1/m) * np.sum((y_pred - y) * x)
    dj_db = (1/m) * np.sum(y_pred - y)

    return dj_dw, dj_db
```

```python
dj_dw, dj_db = compute_gradient(w, x, b, y)

print(f"dj_dw: {dj_dw}")
print(f"dj_db: {dj_db}")
```

```python
alpha = 0.0000001

iterations = 200

cost_history = []

for _ in range(iterations):

    dj_dw, dj_db = compute_gradient(w, x, b, y)

    w = w - (alpha * dj_dw)
    b = b - (alpha * dj_db)

    cost = compute_cost(w, x, b, y)

    cost_history.append(w)

print(f"Final w: {w}")
print(f"Final b: {b}")
```

```python
cost = compute_cost(w, x, b, y)
print(cost)
```

```python
y_pred = predict(w, x, b)

plt.scatter(x, y)
plt.plot(x, y_pred)

plt.xlabel("House Size (sqft)")
plt.ylabel("Price (lakhs)")
plt.title("House Price Prediction")

plt.show()
```

```python
size = 1200

predicted_price = predict(w, size, b)

print("Predicted price:", predicted_price)
```

```python
plt.plot(cost_history)

plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Cost vs Iterations")

plt.show()
```
